```python
import matplotlib.pyplot as plt
import skrf as rf
import numpy
import math

file1='core1.s1p'
od = 25
id = 15
width = 10
aol=width/2/math.pi*math.log(od/id)
turns=11 #use least turns possible to minimise self resonance effects

u0=4e-7*math.pi

nw1=rf.Network(file1)
mu=(nw1.z[:,0,0]/turns**2/(2*math.pi*nw1.f))/(u0*aol)*-1j

plt.figure()
plt.plot(nw1.f/1e6,mu.real,label='mu1')
plt.plot(nw1.f/1e6,-mu.imag,label='mu2')
plt.xscale('log')
plt.legend(('$\mu\'$','$\mu\'\'$'),loc='upper right',shadow=True)
plt.ylim(bottom=0)
plt.ylabel('Relative permeability')
plt.xlabel('Frequency (MHz)')
plt.title('Complex relative permability - {}'.format(file1))
plt.grid(True)
plt.show()

plt.figure()
nw1.plot_s_db()
plt.show()
```
